```python
# Add the essential library matplotlib
import matplotlib.pyplot as plt
import numpy as np
# create the data
a = np.linspace(0, 20)
# Draw the plot
plt.plot(a, a + 0.5, label='linear')
```

```python
x = [1,3,5,7,9,11]
y = [10,25,35,33,41,59]
plt.plot(x, y, label = 'Series 1', color='blue')
x = [2,4,6,8,10,12]
y = [15,29,32,33,38,55]
plt.plot(x, y, label = 'Series 2', color = 'red')
plt.xlabel('X-label')
plt.ylabel('Y-label')
plt.title('Multiple Python Line Graph')
plt.legend()
plt.show()
```

```python
# create the data
x = [1,3,5,7,9,11]
y = [10,25,35,33,41,59]
plt.scatter(x, y, color='red', marker='*', alpha=0.5)
plt.xlabel('X-label')
plt.ylabel('Y-label')
# Add the title to graph
plt.title("Scatter Chart Sample")
# Display the chart
plt.show()
```
